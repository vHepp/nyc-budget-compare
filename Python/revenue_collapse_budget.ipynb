```python
import pandas as pd
import numpy as np
```

```python
date_columns = [
'Publication Date',
'Fiscal Year',
]

catagory_columns = [
'Agency Number',
'Agency Name',
'Fundind Source Name',
# 'Revenue Category',
# 'Revenue Category Name',
# 'Revenue Class Code',
# 'Revenue Class Name',
# 'Revenue Source',
# 'Revenue Source Name',
]

value_columns = [
'Adopted Budget Amount',
'Current Modified Budget Amount',
# 'Year 1 Revenue Amount',
'Year 2 Revenue Amount',
# 'Year 3 Revenue Amount',
# 'Year 4 Revenue Amount',
# 'Year 5 Revenue Amount',
]

target_columns = date_columns + catagory_columns + value_columns
target_columns
```

```python
print(df.columns)
print(df.dtypes)
```

```python
new_df = df[target_columns].copy()

new_df = new_df.sort_values(by=['Agency Number',"Publication Date"], ascending=[True,False])

# df['Fiscal Year'] = df['Fiscal Year'].astype("int64")

new_df["Publication Date"].unique()
```

```python
new_df = new_df[new_df["Publication Date"].isin(['2026 05 12', '2026 02 17'])]
# new_df = new_df.sort_values(by=['Agency Number', 'Revenue Category','Revenue Class Code','Revenue Source',"Publication Date"], ascending=[True,True,True,True,False])

new_df = new_df.reset_index(drop=True)
new_df
```

## Groupings


### Agency By Funding Source ("Fundind Source Name")


```python
base_df = new_df[catagory_columns].drop_duplicates().reset_index(drop=True).copy()

# base_df[value_columns] = 0

base_df
```

```python
grouped_df_1 = new_df.groupby(date_columns + catagory_columns, as_index=False)[value_columns].sum().copy()
grouped_df_1 = grouped_df_1.sort_values(by=['Agency Number', "Fundind Source Name","Publication Date"], ascending=[True,True,False])
grouped_df_1 = grouped_df_1.reset_index(drop=True)

grouped_df_1
```

```python
# for i in grouped_df_1[catagory_columns].value_counts().items():
#     print(i)
```

#### Executive


```python
exec_df_1 = grouped_df_1[grouped_df_1["Publication Date"] == '2026 05 12'].reset_index(drop=True).copy()

exec_df_1 = exec_df_1[catagory_columns + value_columns]

exec_df_1
```

```python
base_df
```

```python
merged_1_df = base_df.merge(exec_df_1, how="left", on=catagory_columns).rename(columns={"Year 2 Revenue Amount":"FY27 Executive Revenue Amount"})
merged_1_df
```

#### Prelim


```python
prelim_df_1 = grouped_df_1[grouped_df_1["Publication Date"] == '2026 02 17'].reset_index(drop=True).copy()
prelim_df_1 = prelim_df_1[catagory_columns + ['Year 2 Revenue Amount']]
prelim_df_1
```

```python
merged_2_df = merged_1_df.merge(prelim_df_1, how="left", on=catagory_columns).rename(columns={"Year 2 Revenue Amount":"FY27 Prelim Revenue Amount"})
merged_2_df
```

```python
merged_2_df.head(10)
```

```python
final_df = merged_2_df.copy()

columns_to_fillna = final_df.columns[-2:]
print(columns_to_fillna)

final_df[columns_to_fillna] = final_df[columns_to_fillna].fillna(0)

final_df.head(10)
```

```python
final_df.to_excel("Docs/Collapsed_Revenues_Agency_by_Funding_Source.xlsx", index=False)
```
